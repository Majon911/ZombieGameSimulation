# The Database

```python
# Connecting to SQL Database
cnx = mysql.connector.connect(
    host='localhost',
    user='root',
    passwd='Baumann1',
    database='ZombieSimulation'
)

cursor = cnx.cursor()
```

```python
# To delete statistics
query = "DROP TABLE statistics"
cursor.execute(query)

for i in cursor:
    print(i)
```

```python
# Creating Table statistics in the database
query = "CREATE TABLE statistics (Time VARCHAR(255), City_Name VARCHAR(255), Healthy INT, Zombies INT, Dead INT, Civil INT, Military INT, Medics INT)"
cursor.execute(query)

query = "SHOW TABLES"
cursor.execute(query)

for i in cursor:
    print(i)
```

```python
# Fetching all of the table
query = "SELECT * FROM statistics"
cursor.execute(query)

rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# Closing the connections and cursor
cursor.close()
cnx.commit()
cnx.close()
```
